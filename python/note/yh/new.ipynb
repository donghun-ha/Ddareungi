## 따릉이 새로 시작

```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
```

```python
df = pd.read_csv("../data/반납추가.csv")
df.head()
```

```python
df_station = pd.read_csv("../data/송파구 대여소 거치대수.csv")
df_station
```

```python
df_station.drop(columns='Unnamed: 0', inplace=True)
df_station
```

```python
df.info()
```

#### 반납된 자치구만 송파인구인것으로 분리

```python
df_new = df[(df['자치구'] == '송파구') & (df['반납 자치구'] == '송파구')]
df_new.head()
```

```python
df_station.info()
```

> df_new와 중복되는 자치구 컬럼 삭제

```python
df_station.drop(columns='자치구', inplace=True)
df_station
```

```python
df_new.columns
```

> df_new와 df_station의 merge

```python
df_merged = df_new.merge(df_station, how='left', left_on='대여 대여소번호', right_on='대여소 번호')
df_merged.head()
```

```python
df_merged.isnull().sum()
```

#### 대여일시, 반납일시를 datetime형식으로 변경

```python
df_merged['대여일시'] = pd.to_datetime(df_merged['대여일시'])
df_merged['반납일시'] = pd.to_datetime(df_merged['반납일시'])

# 데이터 타입 확인
df_merged.info()
```

```python
일별_대여_횟수 = df_merged.groupby(pd.Grouper(key='대여일시', freq='D')).size().reset_index(name='대여횟수')
일별_대여_횟수
```

> 그래프한번 확인

```python
# 한글 폰트 문제 해결 
# matplotlib은 한글 폰트를 지원하지 않음
# os정보
import platform
import matplotlib.pyplot as plt

# font_manager : 폰트 관리 모듈
# rc : 폰트 변경 모듈
from matplotlib import font_manager, rc
# unicode 설정
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic') # os가 macos
elif platform.system() == 'Windows':
    path = 'c:/Windows/Fonts/malgun.ttf' # os가 windows
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print("Unknown System")
```

```python
import matplotlib.pyplot as plt

# 일별 대여 횟수 시각화
plt.figure(figsize=(10, 6))
plt.bar(일별_대여_횟수['대여일시'], 일별_대여_횟수['대여횟수'], color='skyblue')
plt.title('일별 대여 횟수')
plt.xlabel('대여일시')
plt.ylabel('대여횟수')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### 날짜 컬럼 생성하여 년도, 일, 월 분리

```python
df_merged['날짜'] = pd.to_datetime(df_merged['대여일시'], format='%Y/%m/%d %H:%M:%S').dt.date
df_merged.head()
```

```python
df_merged['날짜'] = pd.to_datetime(df_merged['날짜'])
df_merged['연도'] = df_merged['날짜'].dt.year
df_merged['월'] = df_merged['날짜'].dt.month
df_merged['일'] = df_merged['날짜'].dt.day
```

```python
import holidays

# 대한민국 공휴일 불러오기
kr_holidays = holidays.KR(years=range(2020,2025))

# 요일 숫자를 요일 이름으로 변환
df_merged['요일'] = df_merged['날짜'].dt.weekday

# 공휴일 체크
df_merged['휴일여부'] = df_merged['날짜'].apply(lambda x: 2 if x in kr_holidays else (1 if x.weekday() >= 5 else 0))

# 데이터 확인
df_merged
```

# 결측치 확인

```python
missing_values = df_merged.isnull().sum()
print(missing_values[missing_values > 0])
```

> 자전거 구분은 새싹인지 아닌지 구분하는 것 이므로 컬럼삭제

```python
df_merged.drop(columns='자전거구분', inplace=True)
```

```python
# 결측치 확인
import missingno as msno 

msno.matrix(df_merged, figsize=(12, 5))
plt.show()
```

```python
# '성별' 컬럼에서 null 값을 가진 행들 확인
null_data = df_merged[df_merged['생년'].isnull()]

# 결과 출력
print(null_data)
```

```python
# NaN 값 삭제 전 행 수 확인
print("삭제 전 행 수:", len(df_merged))

# 반납대여소번호, 자치구, 반납 자치구의 NaN 값이 있는 행 삭제
df_merged = df_merged.dropna(subset=['반납대여소번호', '자치구', '반납 자치구'])

# 삭제 후 행 수 확인
print("삭제 후 행 수:", len(df_merged))

# 결과 확인
print(df_merged.isnull().sum())
```

> 합치거나 새로 만들면서 중복됬던 컬럼 삭제

```python
df_merged.drop(columns=['자전거번호','날짜', '연도', '월', '일'], inplace=True)
df_merged.head()
```

> 성별컬럼은 우선은 NaN값이 있긴하나 나중에 쓸 수도 있기에 우선은 그냥 유지

### 시간대별 대여 빈도 확인

```python
# '대여일시'에서 시간 추출
df_merged['시간대'] = df_merged['대여일시'].dt.hour

# 시간대별 이용 횟수 계산
시간대별_대여_빈도 = df_merged['시간대'].value_counts().sort_index()
```

### 요일별 이용 확인

```python
# '대여일시'에서 요일 추출 (0: 월요일, 6: 일요일)
df_merged['요일'] = df_merged['대여일시'].dt.dayofweek

# 요일별 이용 횟수 계산
요일별_대여_빈도 = df_merged['요일'].value_counts().sort_index()
```

```python
import seaborn as sns

plt.figure(figsize=(10, 5))
sns.barplot(x=['월', '화', '수', '목', '금', '토', '일'], y=요일별_대여_빈도.values)
plt.title('요일별 대여 빈도')
plt.xlabel('요일')
plt.ylabel('이용 횟수')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.barplot(x=시간대별_대여_빈도.index, y=시간대별_대여_빈도.values)
plt.title('시간대별 대여 빈도')
plt.xlabel('시간대 (hour)')
plt.ylabel('이용 횟수')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

### 상위 20개 대여소별 이용 빈도

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 대여소별 이용 횟수 계산
rental_counts = df['대여 대여소명'].value_counts()

# 상위 20개 대여소 추출
top_20_stations = rental_counts.head(20)

# 막대 그래프로 시각화
plt.figure(figsize=(12, 6))
sns.barplot(x=top_20_stations.index, y=top_20_stations.values)
plt.title('상위 20개 대여소별 이용 빈도')
plt.xlabel('대여소명')
plt.ylabel('이용 횟수')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 결과 출력
print(top_20_stations)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 대여소별 대여 빈도 계산
rental_counts = df_merged['대여 대여소명'].value_counts()

# 대여소별 반납 빈도 계산
return_counts = df_merged['반납대여소명'].value_counts()

# 상위 20개 대여소 선택
top_20_stations = rental_counts.head(20).index

# 대여와 반납 빈도를 하나의 데이터프레임으로 결합
combined_counts = pd.DataFrame({
    '대여 빈도': rental_counts[top_20_stations],
    '반납 빈도': return_counts[top_20_stations]
})

# 시각화
plt.figure(figsize=(15, 8))
combined_counts.plot(kind='bar', width=0.8)
plt.title('상위 20개 대여소의 대여 및 반납 빈도')
plt.xlabel('대여소명')
plt.ylabel('빈도')
plt.legend(['대여', '반납'])
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 결과 출력
print(combined_counts)
```
