```python
# CSV 파일들이 있는 폴더 경로
path = "../../Data/rent/*.csv"

# 모든 CSV 파일을 리스트로 가져옴
all_files = glob.glob(path)

# 파일들을 읽어 리스트에 저장
df_list = [pd.read_csv(file, encoding='CP949') for file in all_files]

# 모든 DataFrame을 하나로 연결
result = pd.concat(df_list, ignore_index=True)
```
